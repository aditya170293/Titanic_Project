```python
#Price Elasticity = Percentage Change in Quantity Demanded / Percentage Change in the Price
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Demand":[20,30,31,33,30,33,35],
                  "Price":[2000,1800,1850,1700,1800,1700,1600]})
```

```python
df
```

```python
df["% Change in Demand"] = df["Demand"].pct_change()
```

```python
df["% Change in Price"] = df["Price"].pct_change()
```

```python
df
```

```python
df["Price Elasticity"] = df['% Change in Demand']/df['% Change in Price']
df
```
